```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('spotify.csv')

print(df.shape)
df.head(5)
```

```python
#get list of columns to get correct name of column to remove (1nd column after index)
df.columns
```

```python
#drop column that is not needed (axis=1 lets pandas know we are dropping a column vs row)
df = df.drop("Unnamed: 0", axis=1)
```

```python
#shows that the column has been removed
df.head(5)
```

```python
#check data types of columns
df.dtypes
```

```python
#check for missing values (false means no missing values) (True means missing values are present)
df.isna()
#sum up the missing values in each column (isnull() can also be used here instead of isna())
df.isna().sum()
```

```python
#view rows with missing values in the 'artists' column
#this row contains all the missing values in the dataset
df.loc[df["artists"].isna()]
```

```python
#check for duplicate rows (True means duplicate row, False means unique row)
#keep=first marks duplicates as True except for the first occurrence
#keep=last marks duplicates as True except for the last occurrence
#keep=False marks all duplicates as True
df.duplicated(keep=False).sum()
```

```python
#view duplicate rows
df.loc[df.duplicated(keep=False)]
```

```python
#all the duplicated rows aren't next to each other, so we will sort by 'track_id' to get them next to each other
df.loc[df.duplicated(keep=False)].sort_values(by=['track_id'])
```

```python
#view rows with duplicate 'track_id' values

df.loc[df['track_id'].duplicated(keep=False)].sort_values(by=['track_id'])
```

```python
#qualitative assessment of artists

df['artists'].value_counts()
```

```python
#quantitative assessment of numerical columns
#bottom of danceability looks like it might have some outliers-significantly lower than 25th percentile
df.describe()
```

```python
#plot danceability to visualize distribution and potential outliers
#shows that it is negatively skewed-tail on the left side
df['danceability'].hist(bins=60)

```

```python
#boxplot to better visualize potential outliers in danceability using seaborn
#right side of box is 75th percentile, left side is 25th percentile, line in middle is median
#whiskers extend to 1.5 times the interquartile range (IQR)
#points outside the whiskers are considered outliers
sns.boxplot(x=df['danceability'])
```

```python
#Correlation matrix to see relationships between numerical columns

df.columns
df_corr = df[['popularity',
       'duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr   
```

```python
#heatmap of correlation matrix to better visualize relationships between numerical columns
#darker colors indicates a negative correlation, lighter colors indicates a positive correlation

fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df_corr, annot=True, cmap='mako')
```

```python
#pairplot to visualize relationships between acousticness and energy
#after seeing the results we're going to look at a sample with few points for better visualization

sns.pairplot(df[['acousticness', 'energy']], height=5, aspect = 1.5)
```

```python
#pairplot with sample of 5000 points for better visualization

df_sample = df[['acousticness', 'energy']].sample(n=5000, random_state=1)
sns.pairplot(df_sample, height=5, aspect = 1.5)
```

```python
df_sample = df[['acousticness', 'energy']].sample(n=5000, random_state=1)
sns.pairplot(df_sample, height=5, aspect = 1.5, diag_kind='kde')
```

```python
df['track_genre'].value_counts()

df_genres = df[df['track_genre'].isin(['pop', 'folk', 'emo', 'punk-rock', 'electronic'])]

df_sample = df_genres.sample(n=5000, random_state=1)
sns.pairplot(df_sample[['track_genre', 'acousticness', 'energy']], height=5, aspect = 1.5, diag_kind='kde', hue='track_genre')

```
